```python
import cv2
from google.colab.patches import cv2_imshow

image_path = '/content/apple image.jpg'

# Read the image was using cv2.imread()
img = cv2.imread(image_path)

# check if the image was loded successfully
if img is None:
   print(f"Error: Image not found at {image_path} or unable to load.")
else:
   # Display the image using cv2_imshow
   cv2_imshow(img)
```
